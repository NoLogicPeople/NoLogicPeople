```python
import os
import torch
import torchaudio
from transformers import AutoTokenizer, AutoModelForTextToWaveform

tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-tur")
model = AutoModelForTextToWaveform.from_pretrained("facebook/mms-tts-tur")

input_dir = "texts"
output_dir = "tts_outputs"
os.makedirs(output_dir, exist_ok=True)

for i in range(1, 11):
    file_path = os.path.join(input_dir, f"text{i}.txt")
    with open(file_path, "r", encoding="utf-8") as f:
        text = f.read().strip()

    inputs = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        waveform = model(**inputs).waveform.squeeze(0)

    output_path = os.path.join(output_dir, f"output_{i}.wav")
    torchaudio.save(output_path, waveform.unsqueeze(0), sample_rate=16000)
```
